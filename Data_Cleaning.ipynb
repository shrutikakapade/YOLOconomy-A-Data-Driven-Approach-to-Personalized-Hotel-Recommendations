```python
#Drop the columns that we dont needed
data.drop("Unnamed: 0" ,axis = 1 ,inplace =True)
```

```python
data["Property_name"] = data["Property_name"].str.lower()
data["Location"] =data["Location"].str.lower()
data["Review"] =data["Review"].str.lower()
data["Type_of_stay"] =data["Type_of_stay"].str.lower()
```

```python
data["price"] =data["price"].replace("[\D]" ,"",regex =True).astype(float)
data["Taxes_and_charges_inr"] =data["Taxes_and_charges_inr"].replace("[\D]" ,"",regex =True).astype(float)
```

```python
data.info()
```

```python
data.Taxes_and_charges_inr.unique()
```

```python
data['Rating'].fillna(data['Rating'].median(), inplace=True)
data['Taxes_and_charges_inr'].fillna(0, inplace=True)
data['Review'].fillna('No Review', inplace=True)
```

```python
data.isna().sum()
```

```python
data.dropna(inplace =True)
```

```python
data.isna().sum()
```

```python
#here we extract the city name by using location column
data["City"] = data["Location"].apply(lambda x: x.split(",")[-1].strip() if isinstance(x, str) else None)
```

```python
'''Stay_duration_and_group_size: by using this column i create a 2 more column for the number of adults and children 
to make better understanding about data so for that i use lambda function '''

x = lambda x : x.str.split(" ").str[2]
data[["No_Of_Adults"]] =data[["Stay_duration_and_group_size"]].apply(x).astype(int)
y = lambda x : x.str.split(" ").str[4]
data[["No_Of_Children"]] =data[["Stay_duration_and_group_size"]].apply(y)
```

```python
data
```

```python
#so this column having some null values so i replace them with 0
data[["No_Of_Children"]] = data[["No_Of_Children"]].fillna(0).astype(int)
```

```python
#Drop stay_duration_And_group_size column
data.drop("Stay_duration_and_group_size",axis = 1 ,inplace = True)
```

```python
#create a new column Final_Price by using the Original_price and Taxes_and_charges_inr columns 
data["Final_Price"] =data["price"]+data["Taxes_and_charges_inr"]
```

```python
data
```

```python
#here I create a new column recommended_for
recommend = lambda x : x==0
data["Recommended for"] = data["No_Of_Children"].apply(recommend).astype(str)
```

```python
data["Recommended for"] =data["Recommended for"].replace("True","Recommended for Friends Trip").str.lower()
data["Recommended for"] =data["Recommended for"].replace("false","Recommended for Family Trip").str.lower()
```
